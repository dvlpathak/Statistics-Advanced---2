```python
Q1. What is hypothesis testing in statistics?

Answer:
    Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data.

We start with an assumption (the null hypothesis, H₀).
We collect sample data and use a test statistic to see if there is enough evidence to reject H₀ in favor of the alternative hypothesis, H₁.
It helps answer questions like, “Is a new drug more effective than the old one?” by comparing sample results to a probability model.

Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:
    Null Hypothesis (H₀)

Assumes no effect or no difference

It's the default or starting assumption

We test this hypothesis

Example: "The new drug has no effect"

    Alternative Hypothesis (H₁ or Ha)

Assumes an effect or a difference exists

What the researcher wants to prove

Accepted if H₀ is rejected

Example: "The new drug works better"

Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer:
    The significance level is the probability of rejecting H₀ when it is actually true.

Common values: 0.05 (5%), 0.01 (1%).

It sets the threshold for deciding whether observed data is “rare enough” to reject H₀.
Example: α = 0.05 means we accept a 5% chance of making a wrong rejection.

Q4 . What are Type I and Type II errors? Give examples of each.

Answer:
    Type I Error (False Positive)

Definition: Rejecting the null hypothesis (H₀) when it is actually true.

Meaning: You think there is an effect or difference, but there isn’t.

Example:
A COVID-19 test says a healthy person has the virus.
(You wrongly detect a disease that isn't there.)

    Type II Error (False Negative)

Definition: Failing to reject the null hypothesis when it is actually false.

Meaning: You think there is no effect, but there actually is.

Example:
A COVID-19 test says an infected person is healthy.
(You miss a real disease.)

Q5. What is the difference between a Z-test and a T-test? Explain when to use each.

Answer:
     Z-test

Used when the population standard deviation (σ) is known

Sample size is usually large (n > 30)

Assumes the data follows a normal distribution or large enough for Central Limit Theorem to apply

Test statistic follows a standard normal distribution (Z-distribution)

Example: Testing the average height of a large group when σ is known

     T-test

Used when the population standard deviation (σ) is unknown

Sample size is usually small (n < 30)

Uses the sample standard deviation (s) instead of σ

Test statistic follows a t-distribution, which has thicker tails to account for extra uncertainty

Example: Testing the average weight of a small sample where σ is unknown


Q6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

Answer:
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n, p = 10, 0.5
data = np.random.binomial(n, p, 1000)

plt.hist(data, bins=range(n+2), edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()


Q7. Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

Answer:
import numpy as np
from scipy.stats import norm

data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

# Hypotheses: H0: μ = 50, H1: μ ≠ 50
z = (mean - 50) / (std/np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Mean: {mean:.2f}, Z: {z:.2f}, p-value: {p_value:.4f}")

Q8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

Answer:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(loc=100, scale=15, size=1000)
mean, std = np.mean(data), np.std(data, ddof=1)
ci = norm.interval(0.95, loc=mean, scale=std/np.sqrt(len(data)))

print(f"Mean: {mean:.2f}, 95% CI: {ci}")

plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()


Q9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.

Answer:
import numpy as np
import matplotlib.pyplot as plt

def z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

sample = np.random.normal(50, 10, 1000)
z = z_scores(sample)

plt.hist(z, bins=30, edgecolor='black', alpha=0.7)
plt.title("Z-Score Standardized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
    

                                                             
```
